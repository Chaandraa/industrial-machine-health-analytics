# 06 Machine Learning - Model Building

**Objective**: Train machine learning models (Logistic Regression, Random Forest, and XGBoost) to classify equipment failure risk.

**Inputs**:
- `data/processed/processed_v1.csv`

**Outputs**:
- `models/logistic_regression_v1.joblib`
- `models/random_forest_v1.joblib`
- `models/xgboost_v1.joblib`

```
import pandas as pd
import numpy as np
import joblib
import os
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
from src.config import PROCESSED_DATA_PATH, MODELS_DIR, RANDOM_STATE, TEST_SIZE

# Load feature dataset
df = pd.read_csv(PROCESSED_DATA_PATH)
features = [
    'Ambient_Room_Temp_C', 'Internal_Device_Temp_C', 
    'Cooling_Fan_Speed_RPM', 'Motor_Load_Nm', 
    'Operating_Hours', 'Temp_Diff', 'Utilization_Pct', 'Failure_Risk_Index'
]
X = df[features]
y = df['Device_Failure']

# Stratified train-test split due to target imbalance
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=y
)
print(f'Train shape: {X_train.shape}, Test shape: {X_test.shape}')
```

```
# 1. Train Logistic Regression (Baseline)
lr = LogisticRegression(max_iter=1000, random_state=RANDOM_STATE)
lr.fit(X_train, y_train)
joblib.dump(lr, MODELS_DIR / 'logistic_regression_v1.joblib')
print('Logistic Regression trained and saved.')
```

```
# 2. Train Random Forest (Interpretable Non-linear Classifier)
rf = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=RANDOM_STATE, class_weight='balanced')
rf.fit(X_train, y_train)
joblib.dump(rf, MODELS_DIR / 'random_forest_v1.joblib')
print('Random Forest trained and saved.')
```

```
# 3. Train XGBoost (Comparison Benchmark)
xgb_model = xgb.XGBClassifier(n_estimators=100, max_depth=5, learning_rate=0.1, random_state=RANDOM_STATE, eval_metric='logloss')
xgb_model.fit(X_train, y_train)
joblib.dump(xgb_model, MODELS_DIR / 'xgboost_v1.joblib')
print('XGBoost model trained and saved.')
```

## Key Findings
- Stratified partitioning successfully preserved the 3.39% failure rate across train and test sets.
- Three models (Logistic Regression, Random Forest, and XGBoost) have been successfully compiled and cached.
