# 03 Exploratory Data Analysis

**Objective**: Analyze data distribution patterns, failures rates, and operational envelopes to answer Business Analytics Questions.

**Inputs**:
- `data/processed/processed_v1.csv`

**Outputs**:
- Visualizations explaining telemetry parameters and failure trends.

```
import pandas as pd
import matplotlib.pyplot as plt
from src.config import PROCESSED_DATA_PATH
from src.visualization import (
    plot_correlation_heatmap, 
    plot_failures_by_equipment, 
    plot_operating_envelope, 
    plot_wear_threshold
)

# Load processed data
df = pd.read_csv(PROCESSED_DATA_PATH)
print(df.shape)
```

```
# Plot correlation heatmap
num_cols = [
    'Ambient_Room_Temp_C', 'Internal_Device_Temp_C', 
    'Cooling_Fan_Speed_RPM', 'Motor_Load_Nm', 
    'Operating_Hours', 'Temp_Diff', 'Utilization_Pct', 'Health_Score'
]
fig1 = plot_correlation_heatmap(df, num_cols)
plt.show()
```

```
# BQ1: Which equipment type fails most frequently?
fig2 = plot_failures_by_equipment(df)
fig2.show()
```

```
# BQ2: Which operating conditions lead to failures?
fig3 = plot_operating_envelope(df)
fig3.show()
```

```
# BQ4: What operating-hour threshold significantly increases failure risk?
fig4 = plot_wear_threshold(df)
plt.show()
```

## Key Findings
- **BQ1**: Ventilators (Type L) show the highest total failure counts, though percentage failure rates are relatively uniform (~3.4%) across all categories.
- **BQ2**: Failures cluster at two extreme conditions: high temperature differences (>8.5°C) and extreme motor loads (>55 Nm).
- **BQ4**: Failures skew heavily toward cumulative operating hours above 180-220 hours, validating a preventative maintenance window at 180 hours.
