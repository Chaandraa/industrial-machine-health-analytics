# 02 Data Cleaning & Audit

**Objective**: Run the Data Quality Report suite to validate bounds, map industrial names to healthcare titles, and save the intermediate `cleaned_v1.csv`.

**Inputs**:
- `data/raw/predictive_maintainance.csv`

**Outputs**:
- `data/cleaned/cleaned_v1.csv`
- `data/cleaned/data_quality_report_v1.json`

```python
import pandas as pd
from src.data_loader import load_raw_telemetry
from src.preprocessing import run_data_quality_report, rename_and_map_columns, convert_units_and_clean
from src.config import CLEANED_DATA_PATH

# 1. Ingest
raw_df = load_raw_telemetry()

# 2. Run Data Quality Audit
dq_report = run_data_quality_report(raw_df)
print(f'Completeness: {dq_report["completeness_percentage"]}%')
print(f'Duplicates: {dq_report["duplicates"]}')
print(f'Consistency Checks: {dq_report["consistency_checks"]}')
```

```python
# 3. Map and Clean
mapped_df = rename_and_map_columns(raw_df)
cleaned_df = convert_units_and_clean(mapped_df)
print(f'Cleaned shape: {cleaned_df.shape}')
cleaned_df.head()
```

```python
# Save Cleaned dataset
cleaned_df.to_csv(CLEANED_DATA_PATH, index=False)
print(f'Saved Cleaned CSV to: {CLEANED_DATA_PATH}')
```

## Key Findings
- Data completeness is 100% (no missing values).
- Duplicate row checks return 0, proving row uniqueness.
- Renamed columns map to healthcare equipment terminology. Temperature units converted to Celsius.
